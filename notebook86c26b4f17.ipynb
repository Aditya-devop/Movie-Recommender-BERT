# BERT Model: Movie Recommender

```python
import pandas as pd
import ast
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load datasets
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
movies = pd.read_csv('/content/tmdb_5000_movies.csv')

# Merge credits and movies
movies = movies.merge(credits, on='title')
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew', 'production_companies']]
movies.dropna(inplace=True)

# Helper functions
def convert(text):
    return [i['name'].replace(" ", "") for i in ast.literal_eval(text)]

def fetch_director(text):
    return [i['name'].replace(" ", "") for i in ast.literal_eval(text) if i['job'] == 'Director']

def process_list_column(col):
    return movies[col].apply(convert)

movies['genres'] = process_list_column('genres')
movies['keywords'] = process_list_column('keywords')
movies['cast'] = movies['cast'].apply(lambda x: [i['name'].replace(" ", "") for i in ast.literal_eval(x)[:3]])
movies['crew'] = movies['crew'].apply(fetch_director)
movies['production_companies'] = process_list_column('production_companies')
movies['overview'] = movies['overview'].apply(lambda x: x.split())

# Combine all tags
movies['tags'] = (
    movies['overview'] + movies['genres'] + movies['keywords'] +
    movies['cast'] + movies['crew'] + movies['production_companies']
)
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))

# Keep relevant columns
new = movies[['movie_id', 'title', 'tags']]

# Load BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Get embeddings for all movie tags
movie_embeddings = model.encode(new['tags'].values, show_progress_bar=True)

# Semantic search function
def recommend(query, top_n=5):
    query_embedding = model.encode([query])[0]
    similarities = cosine_similarity([query_embedding], movie_embeddings)[0]
    movie_indices = similarities.argsort()[-top_n-1:][::-1][1:]

    print(f"🔍 Top {top_n} matches for: \"{query}\"\n")
    for idx in movie_indices:
        print(new.iloc[idx].title)

# Examples
# recommend('space sci-fi adventure')            # by genre/theme
# recommend('Leonardo DiCaprio')                # by cast
# recommend('Warner Bros')                      # by production company
# recommend('James Cameron director')           # by crew
# recommend('In the 22nd century')              # by overview theme
```
